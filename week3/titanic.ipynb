```python
import pandas as pd

# Titanicデータセットを読み込む
titanic_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tomoshibi/titanic.csv') # 先ほど取得したpathに書き換えて下さい
titanic_df
```

```python
# 男性の生存率を求める
dfm = titanic_df[titanic_df['Sex'] == 'male']
male_survival_rate = dfm["Survived"].mean()

# 女性の生存率を求める
dff = titanic_df[titanic_df['Sex'] == 'female']
female_survival_rate = dff["Survived"].mean()


print(f"男性の生存率: {male_survival_rate:.3f}")
print(f"女性の生存率: {female_survival_rate:.3f}")
```

```python
import matplotlib.pyplot as plt
titanic_df.Age.hist(bins=50, figsize=(12, 8))
plt.show
```

```python
age_sur = pd.Series(index=range(81), dtype= 'float')
age = 0
while age <= 80.0:
  age_sur[age] = titanic_df[titanic_df['Age'] == age]['Survived'].mean()
  age += 1
print(age_sur)
age_sur.plot(kind='bar',figsize=(15,6))
plt.title("Survival rate by age")
plt.xlabel("Age")
plt.ylabel("Survival rate")
plt.show()
```
